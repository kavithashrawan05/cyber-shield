```python
import matplotlib.pyplot as plt

# Make sure folder exists
os.makedirs('outputs', exist_ok=True)

# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
incidents = [2, 5, 1, 3, 4]

plt.bar(days, incidents, color='red')
plt.title('Ransomware Incidents This Week')
plt.xlabel('Day')
plt.ylabel('Number of Attacks')
plt.savefig('outputs/ransomware_chart.png')
plt.show()
```
